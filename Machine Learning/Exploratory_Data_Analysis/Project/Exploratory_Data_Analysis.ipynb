<h2 style="color:Blue;"><b>I. Data Description</b></h2>

* Age : Age of the patient
* Sex : Sex of the patient
* exang: exercise induced angina (1 = yes; 0 = no)
* ca: number of major vessels (0-3)
* cp : Chest Pain type chest pain type
            Value 1: typical angina
            Value 2: atypical angina
            Value 3: non-anginal pain
            Value 4: asymptomatic
* trtbps : resting blood pressure (in mm Hg)
* chol : cholestoral in mg/dl fetched via BMI sensor
* fbs : (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
* rest_ecg : resting electrocardiographic results
            Value 0: normal
            Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or       depression of > 0.05 mV)  
            Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
* thalach : maximum heart rate achieved
* target : 0= less chance of heart attack 1= more chance of heart attack. </b>

```python
# Checking outliers
numerical = ['age','trtbps','chol','thalachh','oldpeak']

plt.figure(figsize=(8,6))
plt.title("Numerical Variables Outliers")
ax = sns.boxplot(data = data[numerical])
```

```python
# Removing outliers
for i in data[numerical]:
    Q1 = data[i].quantile(0.25)
    Q3 = data[i].quantile(0.75)
    interQ = Q3-Q1
    LowerT = Q1 - 1.5 * interQ
    UpperT = Q3 + 1.5 * interQ
    median = np.median(data[i])
    for j in data[i]:
        if j > UpperT or j < LowerT:
            data[i] = data[i].replace(j, median)
```

```python
plt.figure(figsize=(8, 6))
plt.title("Numerical Variables without Outliers")
ax = sns.boxplot(data = data[numerical])
```

#### Polynomial Features

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree = 2)
```

```python
features = ['chol', 'trtbps'] #cholestoral and Blood pressure
poly_features = pf.fit(X[features])
```

```python
featured = pd.DataFrame(poly_features.transform(X[features]), 
                        columns = poly_features.get_feature_names(input_features = features))
new_cols = []
for i in featured.columns[3:]:
    new_cols.append(i)
```

```python
X_featured = X.join(featured[new_cols])
X_featured
```

<h2 style="color:Blue;"><b>V. Hypothesis Testing<b></h2>

#### Hypothesis:
- The increase of Cholesterol will increase chance of having Heart attack
- High blood pressure will increase the Heart attack
- High heart rate will increase the chance of heart attack


#### Signficance testing
H0: Cholesterol will increase the Heart Attack.

HA: Cholesterol will not increase the chance of having the heart attack.

```python
from sklearn.linear_model import LinearRegression as LinearReg
from sklearn.feature_selection import f_regression as fReg
import statsmodels.api as sm
```

```python
x = X_featured['chol'].to_numpy().reshape(-1, 1)
y = Y.to_numpy().reshape(-1, )

reg = LinearReg().fit(x, y)

r_sq = reg.score(x, y)
p_val = fReg(x, y)[1][0]

print("The R-squared is {} \nThe p_value is {}".format(r_sq, p_val))
```

### Result

The p value is 0.04378991393694867 and is smaller than 0.05. Therefore H0 is rejected, that the higher Cholestoral will lead to chance of having the heart attack.
